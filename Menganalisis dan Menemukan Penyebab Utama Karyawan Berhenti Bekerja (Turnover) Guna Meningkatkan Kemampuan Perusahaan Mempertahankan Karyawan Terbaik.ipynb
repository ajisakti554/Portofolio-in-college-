**Hubungan Variabel Kategorik dan Atrisi (Stacked/Grouped Bar Charts)**

```python
# Variabel Kategorik Kunci
key_categorical_cols = ['Department', 'JobRole', 'MaritalStatus', 'EducationField', 'BusinessTravel']

plt.figure(figsize=(18, 12))
plt.suptitle('Tingkat Atrisi Berdasarkan Kategori', fontsize=16, y=1.02)

for i, col in enumerate(key_categorical_cols):
    plt.subplot(2, 3, i + 1)

    # Membuat tabel silang (crosstab) antara kolom kategori dan Attrition (target)
    # Menghitung persentase horizontal (across the category)
    crosstab_norm = pd.crosstab(df[col], df['Attrition'], normalize='index') * 100

    # Plotting
    crosstab_norm.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='coolwarm')

    plt.title(f'Atrisi di Setiap {col}', fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Persentase Proporsi (%)')
    plt.legend(title='Attrition', loc='lower left')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("-" * 70)
print("Visualisasi proporsi atrisi berdasarkan kategori telah selesai.")
```

# Agregasi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Map 'Yes' to 1 and 'No' to 0 for numerical analysis
df['Attrition_Numeric'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Atur gaya visualisasi
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100 # Meningkatkan resolusi visualisasi

# Kolom kategori yang ingin dianalisis secara mendalam
key_groups = ['MaritalStatus', 'BusinessTravel', 'JobRole', 'Department']

print("🎯 Tingkat Atrisi (Rata-rata) Berdasarkan Kategori Kunci:")

for col in key_groups:
    # 1. Grouping data dan menghitung rata-rata Atrisi (dikalikan 100 untuk persentase)
    attrition_rate_per_group = df.groupby(col)['Attrition_Numeric'].mean().mul(100).sort_values(ascending=False)

    print(f"\n--- Tingkat Atrisi berdasarkan {col} ---")
    print(attrition_rate_per_group.round(2).astype(str) + '%')

    # 2. Penyesuaian Ukuran Visualisasi (khusus untuk JobRole yang banyak kategori)
    if col == 'JobRole':
        plt.figure(figsize=(12, 6)) # Ukuran lebih besar untuk banyak kategori
        rotation_angle = 60
    else:
        plt.figure(figsize=(8, 5))
        rotation_angle = 30

    # Plotting (menggunakan palet yang menonjolkan risiko)
    # Updated to address FutureWarning: assigning `x` to `hue` and setting `legend=False`
    ax = sns.barplot(x=attrition_rate_per_group.index,
                     y=attrition_rate_per_group.values,
                     hue=attrition_rate_per_group.index, # Assign x to hue
                     palette='Reds_r',
                     legend=False) # Set legend to False

    plt.title(f'Tingkat Atrisi per {col}', fontsize=14)
    plt.ylabel('Tingkat Atrisi (%)', fontsize=12)
    plt.xlabel(col, fontsize=12)
    plt.xticks(rotation=rotation_angle, ha='right')

    # 3. Menambahkan Teks Nilai (Persentase) di atas Setiap Batang
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 9),
                    textcoords = 'offset points',
                    fontsize=10,
                    fontweight='bold')

    plt.tight_layout() # Memastikan label tidak terpotong
    plt.show()

print("-" * 70)
```

**Interpretasi Tingkat Atrisi per marital status:**

 Berdasarkan Status PernikahanVisualisasi ini secara jelas mengidentifikasi Status Lajang ($\text{Single}$) sebagai pendorong risiko atrisi terbesar. Dengan tingkat atrisi mencapai $\text{25.53%}$, karyawan berstatus lajang keluar sekitar dua kali lipat lebih sering daripada karyawan yang Menikah ($\text{12.48%}$) atau Bercerai ($\text{10.06%}$).  Fenomena ini mengisyaratkan bahwa karyawan lajang memiliki ikatan yang lebih longgar—baik secara sosial maupun profesional—yang memungkinkan mereka lebih mudah mencari peluang di tempat lain. Bagi perusahaan, ini berarti program retensi harus segera difokuskan untuk meningkatkan keterikatan sosial dan dukungan keseimbangan hidup yang unik bagi karyawan yang tidak memiliki komitmen keluarga.

```python
# Metrik kunci untuk dibandingkan
metrics_to_compare = ['MonthlyIncome', 'JobSatisfaction', 'PerformanceRating', 'YearsAtCompany', 'WorkLifeBalance']

print("💰 Perbandingan Rata-Rata Metrik Kunci (Atrisi vs. Non-Atrisi):")

# Grouping data berdasarkan status Atrisi (0 dan 1)
comparison_df = df.groupby('Attrition_Numeric')[metrics_to_compare].mean().T

# Mengganti index 0 dan 1 menjadi label yang bermakna
comparison_df.columns = ['No Attrition (0)', 'Yes Attrition (1)']

print(comparison_df.round(2))
print("-" * 70)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Pastikan Anda menggunakan DataFrame 'comparison_df' yang sudah di-transpose (T)

# Mengatur DataFrame untuk visualisasi (melt/unpivot data)
# Kita perlu mengembalikan comparison_df (hasil transpose) ke format 'panjang'
comparison_df_visual = comparison_df.reset_index().melt(id_vars='index',
                                                     var_name='Attrition Status',
                                                     value_name='Average Value')
comparison_df_visual.columns = ['Metric', 'Attrition Status', 'Average Value']


plt.figure(figsize=(14, 8))
plt.suptitle('Perbandingan Rata-Rata Metrik Kunci (Atrisi vs. Non-Atrisi)', fontsize=16)

# Menggunakan Point Plot untuk visualisasi yang rapi dan fokus pada perbedaan
sns.pointplot(data=comparison_df_visual,
              x='Metric',
              y='Average Value',
              hue='Attrition Status',
              dodge=True, # Memisahkan titik
              linestyles=['-', '--'], # Gaya garis berbeda
              markers=['o', 's'], # Gaya marker berbeda
              palette={'No Attrition (0)': 'green', 'Yes Attrition (1)': 'red'}) # Warna yang kontras

plt.xlabel('Metrik', fontsize=12)
plt.ylabel('Nilai Rata-Rata', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Tambahkan grid horizontal

plt.legend(title='Status', loc='best')
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Menyesuaikan layout agar judul tidak terpotong
plt.show()

print("-" * 70)
print("Visualisasi Perbandingan Metrik Kunci Selesai.")
```

**Interpretasi:**

Data pada gambar menunjukkan perbandingan rata-rata metrik utama antara karyawan yang bertahan ("No Attrition") dan yang meninggalkan perusahaan ("Yes Attrition"). Secara sederhana, karyawan yang memilih untuk bertahan di perusahaan cenderung memiliki pendapatan bulanan yang jauh lebih tinggi (rata-rata $6832.74 vs $4787.09), masa kerja yang lebih lama di perusahaan tersebut (7.37 tahun vs 5.13 tahun), serta tingkat kepuasan kerja dan keseimbangan kehidupan-kerja yang sedikit lebih baik. Sementara itu, penilaian kinerja rata-rata hampir sama persis antara kedua kelompok tersebut
